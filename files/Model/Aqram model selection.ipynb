```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Aqram_Project1_FinalClean_data.csv')
```

```python
df.head()
```

```python
X = df[['Lot Frontage','Lot Area','Overall Qual','Year Built','Year Remod/Add','Exter Qual','Bsmt Qual',
        'Bsmt Exposure','BsmtFin SF 1','Total Bsmt SF','1st Flr SF','Gr Liv Area',
        'Kitchen Qual','Fireplace Qu','Garage Finish','Garage Cars',
        'Garage Area','Neighborhood_NridgHt','Sale Condition_Partial']]
y = df['SalePrice']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Elastic

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1, 1, 5, 10, 50, 100],
             'l1_ratio':[.1, .5, .7, .9, .95, .1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator = base_elastic_model,
                         param_grid = param_grid,
                         scoring = 'neg_mean_squared_error',
                         cv = 140,
                         verbose = 2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_estimator_
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
```

```python
MSE
```

```python
RMSE
```

```python
test_res = y_test - y_pred
```

```python
sns.scatterplot(x=y_test, y=test_res)
plt.axhline(y=0, color='r', linestyle='--')
```

```python
sns.displot(test_res, bins=25, kde=True)
```

```python
import scipy as sp
```

```python
fig, ax = plt.subplots(figsize = (10,12), dpi=100)
_ = sp.stats.probplot(test_res, plot=ax)
```
